```python
# STEP 1
import pandas as pd
import mysql.connector
import sqlalchemy as alc
```

```python
# STEP 2
mydb = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    passwd = 'password', # ganti dengan password MySQL
    database = 'world'
)
```

```python
# Kalau mau ganti database yang digunakan, tinggal ganti nama database-nya saja di variable database. Lalu, run cell-nya.
# mydb = mysql.connector.connect(
#     host = 'localhost',
#     user = 'root',
#     passwd = 'password', # ganti dengan password MySQL
#     database = 'sakila'
# )
```

```python
# STEP 3

from sqlalchemy import create_engine

engine = create_engine('mysql+mysqlconnector://root:password@localhost/world') # Ganti kata 'password' dengan password MySQL

# Kalau mau ganti database, run ulang juga engine-nya. Jangan lupa untuk mengganti nama database setelah tanda '/' di akhir kode.
```

```python
# Metode pertama

# Membuat akses database
mycursor = mydb.cursor()

# Tulis query
query = 'SELECT * FROM city'

# Eksekusi query
mycursor.execute(query)

# Simpan hasil query pada variabel bernama 'result'
result = mycursor.fetchall()

# Konversi output menjadi dataframe dan menampilkan hasilnya
df = pd.DataFrame(result, columns = mycursor.column_names)
df.head()
# df.tail()
```

```python
# Metode kedua
# Membuat function untuk mengakses database dan tables

mycursor = mydb.cursor()

def sql_df(your_query):
    mycursor.execute(your_query)
    myresult = mycursor.fetchall()
    df = pd.DataFrame(myresult, columns = mycursor.column_names)
    return df
```

```python
# Syntax untuk memasukkan query

sql_df(
    '''
    SELECT * FROM country
    LIMIT 10;
    '''
)
```

```python
# Metode ketiga

query = '''
SELECT * FROM city
LIMIT 5;
'''

pd.read_sql(query, engine)
```
