**Buatlah DataFrame baru bernama df_top15 yang isinya adalah top 15 negara berdasarkan total imigrannya!**

```python
# DataFrame baru berisi top 15 negara.
df_top15 = df.sort_values(['Total'], ascending=False, axis=0).head(15)
df_top15
```

**Buatlah DataFrame baru bernama df_decades yang isinya adalah aggregasi tiap dekade (80an, 90an, 2000an).**

- 80's: 1980-1989
- 90's: 1990-1999
- 00's: 2000-2009


```python
# List untu tiap dekade (80's, 90's, and 00's).
years_80s = list(map(str, range(1980, 1990))) 
years_90s = list(map(str, range(1990, 2000))) 
years_00s = list(map(str, range(2000, 2010))) 

# Slice df_top15 untuk mengambil data sesuai dengan rentang dekadenya.
df_80s = df_top15.loc[:, years_80s].sum(axis=1) 
df_90s = df_top15.loc[:, years_90s].sum(axis=1) 
df_00s = df_top15.loc[:, years_00s].sum(axis=1)

# Merge 3 slicing di atas menjadi sebuah DataFrame baru.
df_decades = pd.DataFrame({'1980s': df_80s, '1990s': df_90s, '2000s':df_00s}) 

# Menampilkan hasil DataFrame baru.
df_decades.head()
```

```python
# Descriptive stats df_decades
df_decades.describe()
```

**Buatlah boxplot untuk ketiga dekade yang telah dibuat, interpretasikan hasilnya!**

```python
# Boxplot untuk tiap dekade
df_decades.plot(kind='box', figsize=(10, 6))
plt.title('Immigration from top 15 countries for decades 80s, 90s and 2000s');
```

**Negara mana saja yang termasuk ke dalam outliers untuk total imigrasi pada dekade 2000an?**

```python
# Function to detect outliers
def find_anomalies(data):
    anomalies = []
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f'IQR: {iqr}')

    limit_bawah = q1 - limit
    limit_atas = q3 + limit
    print(f'Limit bawah: {limit_bawah}')
    print(f'Limit atas: {limit_atas}')

    for outlier in data:
        if outlier < limit_bawah or outlier > limit_atas:
            anomalies.append(outlier)
            counter = (len(anomalies))
    print(f'Jumlah outliers: {counter}')
    # print(f'anomalies: {anomalies}')        
    
    # return anomalies
```
