### **Line Plot Exercise**

```python
# Import libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
bbni = pd.read_csv('BBNI.JK.csv')
bbri = pd.read_csv('BBRI.Jk.csv')
bmri = pd.read_csv('BMRI.JK.csv')
btpn = pd.read_csv('BTPN.JK.csv')
bbca = pd.read_csv('BBCA.JK.csv')
```

```python
# Mengubah format tipe data variabel 'Date' dari str menjadi datetime
bbni['Date'] = pd.to_datetime(bbni['Date'])
```

**Plot harga buka saham bank tiap harinya dalam interval 1 tahun**

```python
# Variabel untuk sumbu x dan y. Referensi sumbu x adalah 'Date' yang sudah dikonversi menjadi datetime type.
x = bbni['Date']
y1 = bbni['Open']
y2 = bbri['Open']
y3 = bmri['Open']
y4 = btpn['Open']
y5 = bbca['Open']

# Plot graph
plt.figure(figsize=(12, 6))
plt.style.use('default')
plt.plot(x, y1, 'r--', x, y2, 'b-.', x, y3, 'g', x, y4, 'cyan', x, y5, 'purple')
plt.xlabel('Bulan')
plt.xticks(rotation=45)
plt.ylabel('Harga Buka Saham')
plt.suptitle('Histori Harga Buka Saham 5 Bank Tahun 2021', size=16)
plt.legend(['BBNI', 'BBRI', 'BMRI', 'BTPN', 'BBCA'], bbox_to_anchor=(0., 1, 1., .102), loc='lower right', ncol=5, mode='expand', borderaxespad=0);
```

**Plot harga buka saham bank tiap harinya dalam interval 1 tahun untuk masing-masing bank**

```python
bbni['month'] = bbni['Date'].dt.month_name()
bbni
```

```python
import matplotlib.dates as mdates

x = bbni['Date']
y1 = bbni['Open']
y2 = bbri['Open']
y3 = bmri['Open']
y4 = btpn['Open']
y5 = bbca['Open']

plt.style.use('ggplot')
fig = plt.figure(figsize=(30, 12))
half_year_locator = mdates.MonthLocator(interval=1)


ax1 = fig.add_subplot(231)
ax1.plot(x, y1, 'orange')
ax1.set_title('BBNI')
ax1.xaxis.set_major_locator(half_year_locator)
plt.xticks(rotation=20)

ax2 = fig.add_subplot(232)
ax2.plot(x, y2, 'b')
ax2.set_title('BBRI')
ax2.xaxis.set_major_locator(half_year_locator)
plt.xticks(rotation=20)

ax3 = fig.add_subplot(233)
ax3.plot(x, y3, 'darkblue')
ax3.set_title('BMRI')
ax3.xaxis.set_major_locator(half_year_locator)
plt.xticks(rotation=20)

ax4 = fig.add_subplot(234)
ax4.plot(x, y4, 'hotpink')
ax4.set_title('BTPN')
ax4.xaxis.set_major_locator(half_year_locator)
plt.xticks(rotation=20)

ax5 = fig.add_subplot(235)
ax5.plot(x, y5, 'purple')
ax5.set_title('BBCA')
ax5.xaxis.set_major_locator(half_year_locator)
plt.xticks(rotation=20)

fig.suptitle('Histori Harga Buka Saham 5 Bank Tahun 2021', size=26)
plt.show()
```
