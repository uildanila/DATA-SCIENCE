**Choropleth Map**

```python
df_jakarta = pd.read_csv('pendatang_DKI_Maret_2020.csv')
df_jakarta.head()
```

```python
df_jakarta_kelurahan = df_jakarta[['kelurahan', 'jumlah']]
df_jakarta_kelurahan
```

```python
a_map = folium.Map(location=[-6.1753924,106.8249641], zoom_start=13)
a_map
```

```python
jakarta_geojson = 'jakarta_by_kelurahan_geo.geojson'

jakarta_map = folium.Map(location=[-6.1753924,106.8249641], zoom_start=11)

folium.Choropleth(
    geo_data=jakarta_geojson,
    data=df_jakarta_kelurahan,
    columns=['kelurahan', 'jumlah'],
    key_on='feature.properties.name',
    fill_color='PuBuGn',
    fill_opacity=0.6,
    line_opacity=0.2,
    legend_name='Transmigrasi Jakarta Maret 2020'
).add_to(jakarta_map)

jakarta_map
```
