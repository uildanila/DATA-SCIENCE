```python
# Load dataset 1
df_audi = pd.read_csv(r'D:\Car Datasets\audi.csv')
df_audi.head()
```

```python
# Load dataset 2
df_hyundai = pd.read_csv(r'D:\Car Datasets\hyundi.csv')
df_hyundai.head()
```

```python
# Load dataset 3
df_toyota = pd.read_csv(r'D:\Car Datasets\toyota.csv')
df_toyota.head()
```

**1. Jika Anda perhatikan, nama kolom 'tax(£)' di dataset hyundi, nama kolom tersebut berbeda
dengan di dataset lainnya. Ubah nama kolom tersebut menjadi 'tax' agar sama dengan di dataset
yang lain**

```python
# Jika Anda perhatikan, nama kolom 'tax(£)' di dataset hyundi berbeda dengan di dataset lainnya.
# Ubah nama kolom tersebut menjadi 'tax' agar sama dengan di dataset yang lain.

df_hyundai.rename(columns={'tax(£)':'tax'}, inplace=True)
df_hyundai.head(3)
```

**2. Buat kolom baru bernama 'brand' dengan isi nama brand mobilnya, letakkan pada kolom paling kiri. Misalnya untuk DataFrame audi, isi kolom 'brand' dengan ‘Audi’!**

```python
df_audi.insert(0, 'brand', value='Audi')
df_hyundai.insert(0, 'brand', value='Hyundai')
df_toyota.insert(0, 'brand', value='Toyota')
```

```python
df_toyota.head(3)
```

**3. Gabungkan ketiga DataFrame tersebut menjadi sebuah DataFrame bernama df_car. Atur agar index-nya tidak ada yang berulang.**

```python
df_car = pd.concat([df_audi, df_hyundai, df_toyota], axis=0, ignore_index=True)
df_car
```

**4. Tampilkan rata-rata pajak berdasarkan brand dan transmission. Lalu urutkan dari nilai pajak yang paling tinggi**

```python
tax_avg = df_car.groupby(by=['brand', 'transmission'], as_index=False).mean().sort_values(by='tax', ascending=False)[['brand', 'transmission', 'tax']].reset_index()
tax_avg['year'] = df_car['year']
tax_avg.drop(columns='index', axis=1, inplace=True)
tax_avg
```

**5. Anda diminta untuk mencari mobil Toyota Camry yang harganya di bawah rata-rata harga Camry dan jarak tempuhnya di bawah rata-rata jarak tempuh Camry juga. Tampilkan daftar mobil yang sesuai dengan kriteria tersebut!**

```python
df_car['model'].unique()
```

```python
# Menghapus whitespaces yang tidak diperlukan
df_car['model'] = df_car['model'].str.strip()
```

```python
df_car['model'].unique()
```

```python
# buat dataframe untuk mobil Camry saja
df_camry = df_car[df_car['model'] == 'Camry']
df_camry
```

```python
# Buat kolom baru untuk menandakan apakah mobil tersebut di bawah rata-rata mileage
df_camry['mileage_under'] = df_camry[['mileage']] < df_camry[['mileage']].mean()
df_camry
```

```python
# Buat kolom baru untuk menandakan apakah mobil tersebut di bawah rata-rata harga
df_camry['price_under'] = df_camry[['price']] < df_camry[['price']].mean()
df_camry
```

```python
# Kondisi yang keduanya True
df_camry[(df_camry['mileage_under'] == True) & (df_camry['price_under'] == True)]
```

```python
# Cara lain
df_car[(df_car['model'] == 'Camry') & (df_car['mileage'] < df_car.groupby('model').mean().loc['Camry', 'mileage']) & (df_car['price'] < df_car.groupby('model').mean().loc['Camry', 'price'])]
```

**6. Tampilkan rata-rata jarak tempuh berdasarkan tahun, di mana rata-rata jarak tempuh pada tahun tersebut di bawah median dari seluruh rata-rata jarak tempuh. Urutkan dari rata-rata jarak tempuh yang paling kecil!**

```python
avg_mpy = df_car.groupby('year').mean()[['mileage']]
avg_mpy[avg_mpy < avg_mpy.median()].sort_values('mileage')
```

**7. Tampilkan seluruh informasi dari mobil termahal yang ada di dataset!**

```python
# Mobil termahal
df_car.loc[[df_car['price'].idxmax()]]
```

**8. Mobil Toyota keluaran tahun berapa yang paling banyak datanya pada dataset?**

```python
# Mobil Toyota paling banyak yang keluaran tahun berapa
x = df_car[df_car['brand']=='Toyota']['year'].value_counts()
x.head(1)
```

**9. Tampilkan tipe mobil Toyota keluaran tahun 2019 dan urutkan berdasarkan rata-rata jarak tempuh tertinggi**

```python
df_2019 = df_car[df_car['year']==2019]

df_2019.groupby(['brand','model']).mean().loc['Toyota'].sort_values(by='mileage', ascending=False)[['mileage','mpg']]
```

**10. Tampilkan jumlah mobil Audi keluaran 2019 berdasarkan model**

```python
# Jumlah mobil Audi keluaran 2019 berdasarkan model
df_car[(df_car['brand']=='Audi') & (df_car['year']==2019)][['model']].value_counts()
```
