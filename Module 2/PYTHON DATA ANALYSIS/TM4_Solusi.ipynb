**20. Buatlah kolom baru yang isinya adalah income category (low, medium, high) berdasarkan variabel penghasilan tahunan**

**23. Buatlah DataFrame untuk menampilkan nilai rata-rata dari penghasilan tahunan, resensi, serta jumlah pembelian pada toko dan online berdasarkan kelompok pada kolom level pendidikan**

```python
# Buatlah DataFrame untuk menampilkan nilai rata-rata dari income, recency, jumlah pembelian pada toko dan online berdasarkan kelompok pada educational level!

agregat = pd.DataFrame(df.groupby(by='educational_level')[['annual_income', 'recency', 'store_purchases', 'online_purchases']].mean()).reset_index()
agregat
```

```python
# Contoh kalau index tidak di-reset
agregat2 = pd.DataFrame(df.groupby(by='educational_level')[['annual_income', 'recency', 'store_purchases', 'online_purchases']].mean())
agregat2
```

**24. Buatlah pivot table untuk menampilkan data jumlah pembelian dan rata-rata resensi berdasarkan kelompok pada kolom level pendidikan dan status pernikahan**

```python
# Buatlah pivot table untuk menampilkan data jumlah pembelian dan rata-rata recency berdasarkan kelompok pada education level dan marital status!

pivot_table = pd.DataFrame(pd.pivot_table(df, 
                                            values=['sum_purchases', 'recency'], 
                                            index=['marital_status'],
                                            columns=['educational_level'], 
                                            aggfunc={'recency': np.mean, 'sum_purchases': np.sum}, 
                                            fill_value=0)).reset_index()
pivot_table
```

```python
# Contoh tanpa reset_index()
pivot_table2 = pd.DataFrame(pd.pivot_table(df, 
                                            values=['sum_purchases', 'recency'], 
                                            index=['marital_status'],
                                            columns=['educational_level'], 
                                            aggfunc={'recency': np.mean, 'sum_purchases': np.sum}, 
                                            fill_value=0))
pivot_table2
```
