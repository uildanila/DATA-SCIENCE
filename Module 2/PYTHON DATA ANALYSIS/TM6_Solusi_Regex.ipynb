**1. Buatlah function untuk mengecek apakah sebuah string hanya mengandung alphanumeric atau bukan!**

```python
def karakter_khusus(string):
    char_compile = re.compile(r'[^a-zA-Z0-9.]')
    string = char_compile.search(string)
    return not bool(string)

print(karakter_khusus('ABCDEFabcdef123450')) 
print(karakter_khusus('*&%@#!}{'))
```

**2. Buatlah function yang dapat mencocokkan apakah sebuah string mengandung huruf 'a' dan diikuti oleh huruf 'b' sebanyak 0 atau lebih repetisi!**

```python
def text_match(text):
        patterns = 'ab*?'
        if re.search(patterns, text):
                return True
        else:
                return False

print(text_match('ac'))
print(text_match('abc'))
print(text_match('abbc'))
```

**3. Buatlah function yang dapat mencocokkan apakah string memiliki huruf 'm' dan diikuti oleh 3 huruf 'n'!**

```python
def text_match(text):
        patterns = 'mn{3}?'
        if re.search(patterns, text):
                return True
        else:
                return False

print(text_match('mnnn'))
print(text_match('mmnno'))
```

**4. Buatlah function yang dapat mencocokkan apakah string memiliki huruf 'r' dan diikuti oleh 2 hingga 3 huruf 's'!**

```python
def text_match(text):
        patterns = 'rs{2,3}?'
        if re.search(patterns, text):
                return True
        else:
                return False

print(text_match('rs'))
print(text_match('rrssssst'))
```

**5. Buatlah function untuk dapat mencocokkan rangkaian kata dengan huruf kecil yang disambungkan dengan underscore!**

```python
def text_match(text):
        patterns = '^[a-z]+_[a-z]+$'
        if re.search(patterns, text):
                return True
        else:
                return False

print(text_match('aab_cbbbc'))
print(text_match('aab_Abbbc'))
print(text_match('Aaab_abbbc'))
```

**6. Buatlah function yang dapat mencocokkan sebuah string memiliki huruf 'a' diikuti oleh karakter apapun dan karakter terakhirnya adalah huruf 'b'!**

```python
def text_match(text):
        patterns = 'a.*?b$'
        if re.search(patterns, text):
                return True
        else:
                return False

print(text_match('aabbbbd'))
print(text_match('aabAbbbc'))
print(text_match('accddbbjjjb'))
```

**7. Buatlah function yang dapat mencocokkan bahwa sebuah string dimulai dengan sebuah kata!**

```python
def text_match(text):
        patterns = '^\w+'
        if re.search(patterns, text):
                return True
        else:
                return False

print(text_match('Kemarin paman datang membawa banyak bingkisan.'))
print(text_match(' Kemarin paman datang membawa banyak bingkisan.'))
```

**8. Buatlah function yang dapat mencocokkan bahwa suatu string mengandung huruf 'x'!**

```python
def text_match(text):
        patterns = '\w*x.\w*'
        if re.search(patterns, text):
                return True
        else:
                return False

print(text_match('Latihan regex.'))
print(text_match('Latihan Python.'))
```

**9. Buatlah function yang dapat mencocokkan bahwa sebuah string dimulai dengan angka 3!**

```python
def match_num(string):
    text = re.compile(r'^3')
    if text.match(string):
        return True
    else:
        return False
        
print(match_num('5-9080704'))
print(match_num('3-9080704'))
```

**10. Buatlah function yang dapat menghapus angka 0 pada IP address berikut!**

```python
# Menghapus 0 dengan pola seperti contoh target string
ip = '123.01.044.136'
string = re.sub('\.[0]*', '.', ip)
print(string)
```

```python
# Cara untuk menghapus semua 0
ip = '103.01.044.136'
string = re.sub('[0]', '', ip)
print(string)
```

**11. Buatlah function untuk mencari sebuah kata spesifik dalam sebuah string. Selain itu, tampilkan juga lokasi (index) di mana pattern tersebut ditemukan!**

- Teks: 'Semua anak kucing itu berbulu putih hitam.'
- Kata yang dicari: 'kucing'

```python
pattern = 'kucing'
teks = 'Semua anak kucing itu berbulu putih hitam.'
match = re.search(pattern, teks)
index_awal = match.start()
index_akhir = match.end()
# index_span = match.span()

print(f"Kata '{match.re.pattern}' ditemukan pada '{match.string}', pada index {index_awal} hingga {index_akhir}")
```

**12. Buatlah function untuk mengonversi format date dari yyyy-mm-dd ke format dd-mm-yyyy!**

```python
def ganti_format(dt):
        return re.sub(r'(\d{4})-(\d{1,2})-(\d{1,2})', '\\3-\\2-\\1', dt)
date1 = '2022-02-20'

print('Format awal YYY-MM-DD:', date1)
print('Format baru DD-MM-YYYY:', ganti_format(date1))
```
