#### **Python for Data Manipulation Exercise Solution**

```python
# Import libraries
import numpy as np
import pandas as pd
```

**1.** 
- Buatlah sebuah DataFrame yang datanya merupakan data random integer
dengan rentang nilai 1 hingga 100. 
- Jumlah kolom yang akan dibuat adalah sebanyak 5 kolom
dan jumlah barisnya sebanyak 10 baris. 
- Untuk nama kolomnya, berikan nama Kolom_1, Kolom_2, hingga Kolom_5, lalu untuk barisnya biarkan saja dengan index default. 
- Simpan DataFrame tersebut ke dalam sebuah variable bernama df_belajar, kemudian tampilkan DataFrame-nya.

```python
# Soal 1
np.random.seed(2021)

df_belajar = pd.DataFrame(
    data=np.random.randint(1, 100, (10, 5)), 
    columns = ['Kolom_1', 'Kolom_2', 'Kolom_3', 'Kolom_4', 'Kolom_5'])

df_belajar
```

**2.**

Dari DataFrame df_belajar, coba tampilkan data yang berada pada baris index 0 sampai index 5 dan berada pada Kolom_1 dan Kolom_2. 

```python
# Slicing dengan iloc 
df_belajar.iloc[0:6, 0:2]
```

```python
# Slicing dengan menyebut nama kolomnya
df_belajar[['Kolom_1', 'Kolom_2']][0:6]
```

**3.** 

Dari DataFrame df_belajar, coba tampilkan data yang berada pada baris index nomor 3, 2, 5, 7, 0 dan berada pada kolom dengan nomor Index 1, 3, dan 4.

```python
# Slicing dengan iloc 
df_belajar.iloc[[3, 2, 5, 7, 0], [1, 3, 4]]
```

**4.**

Dari df_belajar, tampilkan data dengan syarat data tersebut nilainya lebih besar dari 10 dan merupakan bilangan genap.

```python
# Indexing dengan kondisi
df_belajar[(df_belajar > 10) & (df_belajar % 2 == 0)]
```

**5.**

Dari df_belajar, buatlah sebuah kolom baru di mana kolom tersebut merupakan hasil modulus dari Kolom_1 dan Kolom_3.

```python
# Buat kolom baru
df_belajar['Kolom Baru'] = df_belajar['Kolom_1'] % df_belajar['Kolom_3']
df_belajar 
```

**6.**

Tambahkan sebuah baris baru pada df_belajar yang isinya merupakan data random
normal sebanyak jumlah kolom DataFrame. 

```python
# Soal 6
df_belajar.loc['Baris Baru'] = np.random.randn(6)
df_belajar
```

```python
# Soal 6
df_belajar.loc['Baris Baru'] = np.random.randint(1, 100, (6))
df_belajar
```

**6. Cek apakah ada missing value pada mpg**

**7. Cari rata-rata horsepower berdasarkan asal mobil. Gunakan rata-rata tersebut untuk mengisi missing value pada mpg!**

```python
mpg.groupby('origin').mean()['horsepower']
```

```python
mpg[mpg['horsepower'].isnull()].index.tolist()
```

```python
mpg[mpg['horsepower'].isnull()]
```

```python
mpg.groupby('origin')['horsepower'].mean()
```

```python
mpg['horsepower'] = mpg.groupby('origin')['horsepower'].apply(lambda x: x.fillna(x.mean()))
mpg
```

```python
# Cek hasil imputasi horsepower berdasar asal negara mobil
mpg.loc[[32]]
```

```python
mpg.loc[[330]]
```

```python
mpg.isna().sum()
```

**8. Buat kolom baru bernama horse_cat. isi dengan _lower_ kalau horsepower<=105 dan _upper_ kalau horsepower di atas 105**

```python
# create a list of conditions
conditions = [
    (mpg['horsepower'] <= 105),
    (mpg['horsepower'] > 105)
]

# create a list of the values we want to assign for each condition
values = ['lower', 'upper']

# create a new column and use np.select to assign values to it using the lists
mpg['horsecat'] = np.select(conditions, values)

mpg
```

```python
# Cara lain
mpg['horsecat'] = mpg['horsepower'].apply(lambda x: 'lower' if x <= 105 else 'upper')
```

```python
# Cara lain 2
mpg['horsecat'] = ['lower' if x <= 105 else 'upper' for x in mpg['horsepower']]
```

**9. Berapa jumlah upper horsepower dan lower horsepower berdasarkan asal negara mobil?**

```python
mpg.groupby(['origin', 'horsecat']).count()['mpg']
```
