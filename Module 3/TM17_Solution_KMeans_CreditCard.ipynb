### **Preprocess the data**

**Drop CUST_ID**

This feature is a unique ID of each customer and unnecessary for clustering purposes.

```python
df.drop(['CUST_ID'], axis=1, inplace=True)
```

**Drop missing values of CREDIT_LIMIT**



There are 313 rows with missing values

```python
df.dropna(subset=['CREDIT_LIMIT'], inplace=True)
```

**Impute missing values of MINIMUM_PAYMENTS**

Since this feature is highly skewed, then we can impute the missing values with its median value.

```python
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].median(), inplace=True)
```

### **Scale the data**

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
df_scaled = pd.DataFrame(df_scaled)
df_scaled.columns = df.columns
df_scaled
```

```python
plt.figure(figsize=(20,35))
for i, col in enumerate(df_scaled.columns):
    if df_scaled[col].dtype != 'object':
        ax = plt.subplot(9, 2, i+1)
        sns.kdeplot(df_scaled[col], ax=ax)
        plt.xlabel(col)
        
plt.show()
```

> Standardization does not alter the data distribution. It only changes the scale.

### **Dimensionality reduction (PCA)**

```python
pca = PCA(random_state=0)
pca.fit(df_scaled)
```

```python
sns.set_style('whitegrid')
plt.plot(pca.explained_variance_ratio_.cumsum());
```

```python
# Variance explained with the number of features
pca = PCA(random_state=0)
pca.fit(df_scaled)
features = range(pca.n_components_)

plt.figure(figsize=(8,4))
plt.bar(features[:16], pca.explained_variance_ratio_[:16], color='lightskyblue')
plt.xlabel('PCA Features')
plt.ylabel('Explained Variance')
plt.xticks(features[:16]);
```

```python
pca.explained_variance_ratio_[0:3].sum()
```

```python
pca = PCA(n_components=3, random_state=0)
X = pca.fit_transform(df_scaled)
```

### **K-Means clustering**

**Elbow method**

```python
kmeans_models = [KMeans(n_clusters=k, random_state=0).fit(X) for k in range (1, 10)]
inertia = [model.inertia_ for model in kmeans_models]

plt.plot(range(1, 10), inertia)
plt.title('Elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia');
```

```python
silhoutte_scores = [silhouette_score(X, model.labels_) for model in kmeans_models[2:6]]
plt.plot(range(2, 6), silhoutte_scores, 'bo-')
plt.xticks([2, 3, 4, 5])
plt.title('Silhoutte scores vs Number of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score');
```

**Silhouette method**

```python
kmeans = KMeans(random_state=0)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
silhouette_score(X, labels, metric = 'euclidean', random_state=0)
```

```python
silhouette_scores = []
number_of_cluster = range(2, 11)

for i in number_of_cluster:
  kmeans = KMeans(n_clusters=i, random_state=0)
  kmeans.fit(X)

  labels = kmeans.labels_
  silhouette_scores.append(silhouette_score(X, labels, metric = 'euclidean', random_state=0))
```

```python
plt.figure(figsize=(12,7))

sns.lineplot(number_of_cluster, silhouette_scores)
sns.scatterplot(number_of_cluster, silhouette_scores)

plt.xticks(number_of_cluster)

plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score');
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_
```

```python
pca_df = pd.DataFrame(data=X, columns=['PC1', 'PC2', 'PC3'])
pca_df['labels'] = labels
pca_df.head()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='PC1', y='PC2', hue='labels', data=pca_df, palette='bright');
```

#### **Interpretation**

If we apply PCA for clustering, the result would be more difficult to interpret since a principal component combines all the features that we defined initially. For this case, we got 2 clusters as the best cluster number for n_components = 2.

However, there is much room for improvement to get a better-explained variance ratio of the first two principal components (we only got 47% explained variance from the first two PC). The existence of outliers may affect this result. Therefore, we can improve the result by handling the outliers and maybe only using selected features that we got from the most important feature of each principal component.
